```python
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline
```

```python
sentence = ["The movie was surprisingly good",
            " It was okay, not great, not bad",
            "Terrible acting and worse editing, Dont watch"
            ]
```

```python
#Load the Pretrained sentiment classification Model

classifier = pipeline("sentiment-analysis", model = "textattack/bert-base-uncased-imdb")
classifier(sentence)
```
