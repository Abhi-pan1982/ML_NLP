```python
from transformers import MarianMTModel, MarianTokenizer

class MarianMultiTranslator:
    def __init__(self):
        # Cache to store loaded models and tokenizers to prevent reloading
        self.models = {}
        self.tokenizers = {}

    def get_model_name(self, src_lang, tgt_lang):
        """Constructs the Hugging Face model name based on language codes."""
        return f"Helsinki-NLP/opus-mt-{src_lang}-{tgt_lang}"

    def load_model(self, model_name):
        """Loads and caches the model and tokenizer."""
        if model_name not in self.models:
            print(f"Loading model: {model_name}...")
            try:
                self.tokenizers[model_name] = MarianTokenizer.from_pretrained(model_name)
                self.models[model_name] = MarianMTModel.from_pretrained(model_name)
            except OSError:
                raise ValueError(f"Model '{model_name}' not found. Check if the language pair exists on Hugging Face.")

        return self.tokenizers[model_name], self.models[model_name]

    def translate(self, text, src_lang, tgt_lang):
        """Translates a single string from source to target language."""
        model_name = self.get_model_name(src_lang, tgt_lang)
        tokenizer, model = self.load_model(model_name)

        # Tokenize the input text
        translated_tokens = model.generate(
            **tokenizer(text, return_tensors="pt", padding=True, truncation=True)
        )

        # Decode the generated tokens back to text
        return [tokenizer.decode(t, skip_special_tokens=True) for t in translated_tokens][0]

# --- Usage Example ---
if __name__ == "__main__":
    translator = MarianMultiTranslator()

    text_to_translate = "The quick brown fox jumps over the lazy dog."
    source_language = "en"

    # Target languages: French, Spanish, German, Italian
    target_languages = ["fr", "es", "de", "it"]

    print(f"Original ({source_language}): {text_to_translate}\n")
    print("-" * 40)

    for tgt in target_languages:
        try:
            translation = translator.translate(text_to_translate, source_language, tgt)
            print(f"Translated to {tgt.upper()}: {translation}")
        except ValueError as e:
            print(e)
```

### Translate text from a CSV file

First, let's create a dummy CSV file for demonstration purposes. In a real scenario, you would already have your CSV file.


```python
import pandas as pd

# Create a dummy CSV file
data = {
    'English_Text': [
        "Hello, how are you?",
        "This is a test sentence.",
        "I love machine learning."
    ]
}
df_csv = pd.DataFrame(data)
df_csv.to_csv('texts_to_translate.csv', index=False)

print("Dummy CSV file 'texts_to_translate.csv' created.")
```

Now, let's load this CSV file into a pandas DataFrame and use the `MarianMultiTranslator` to translate the 'English_Text' column into French and add it as a new column.

```python
from transformers import MarianMTModel, MarianTokenizer # Import if not already done in the cell above

# Instantiate the translator
translator = MarianMultiTranslator()

# Load the CSV file
df_to_translate = pd.read_csv('texts_to_translate.csv')

# Define source and target languages
source_language_csv = "en"
target_language_csv = "fr"

# Apply the translation to a column
df_to_translate['French_Translation'] = df_to_translate['English_Text'].apply(
    lambda x: translator.translate(x, source_language_csv, target_language_csv)
)

# Display the DataFrame with translations
display(df_to_translate)
```
